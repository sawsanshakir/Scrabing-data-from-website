```python
    from bs4 import BeautifulSoup
    import requests
    import urllib.request as urllib2
    from csv import writer
    import pandas as pd
```

```python
url = 'https://www.findcourses.co.uk/search/data-analytics?pageSize=500#scroll=4361'
```

```python
page = requests.get(url)
print(page)
print(page.text)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
lists_1 = soup.find_all('div', class_="emg-serp__row emg-serp__row-has-compare")
lists_2 = soup.find_all('div', class_="emg-serp__row emg-serp__row-has-reviews emg-serp__row-has-compare")

data =[ ]
 

for list_ in lists_1:
    course_title = list_.find('a', class_="emg-serp__link").text.replace('\n',' ').split('    ')
       
    print(course_title)
    data.append(course_title)    

    

info_1 = pd.DataFrame(data)
info_1.columns = ['course_title','1', 'course_subtitle', 'course_description','2', 'course_type', 'duration', 'price', 'location', 'course_delivery', '3', '4', '5', '6']
  
info_1 = info_1.drop(['1','2','3', '4', '5', '6'],  axis=1)
print(info_1)
    #thewriter.writerow(info)
info_1.to_csv('housing.csv', index=False)
            
```
